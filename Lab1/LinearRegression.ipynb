```python
s = """2	0,00000036
4	0,00000037
8	0,000000343
16	0,000000385
32	0,000000423
64	0,000000461
128	0,000000613
256	0,000000795
512	0,000001311
1024	0,000002384
2048	0,000003576
4096	0,000005484
8192	0,000008583
16384	0,000011921
32768	0,000024319
65536	0,000046849
131072	0,000082731
262524	0,000134468
524288	0,000348449
1048576	0,00097084"""

s = s.split(	)

n = []
t = []

for i in range(0, len(s), 2):
    n.append(int(s[i]))
    t.append(float(s[i+1].replace(",", ".")))
    
print(n, t)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Output from ptop.c graphically
plt.style.use('seaborn-whitegrid')
plt.plot(n, t, marker='o', color='b')
plt.show()
```

```python
#Ouput points from ptop.c graphically
plt.style.use('seaborn-whitegrid')
plt.scatter(n, t, marker='o', color='b')
plt.show()
```

```python
coef = np.polyfit(n, t, 1)
poly1d_fn = np.poly1d(coef)

plt.plot(n, t, 'bo', n, poly1d_fn(n), '--k')
plt.show()
print("The cost of communication: beta + tau * n")
print("The latency between processor i and j, beta, is ", coef[0])
print("The per-word transfer time (the bandwith), tau, is ", coef[1])
```
